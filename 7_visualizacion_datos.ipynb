```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('top_10_latinoamerica_sql.csv')
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data['Popularidad'],kde=True,bins=30,color="skyblue")
plt.title('Distribucion de popularidad de las canciones')
plt.xlabel('Popularidad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
top_10 = data.sort_values('Popularidad',ascending=False)
top_10.head(5)
```

```python
top_10 = top_10.drop_duplicates(subset='Nombre de la cancion')
top_10
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

bar_plot = sns.barplot(y='Nombre de la cancion',x='Popularidad',data=top_10,palette='viridis')
plt.title('Top 3 canciones mas populares')
plt.xlabel('Nombre de la cancion')
plt.ylabel('Popularidad')
plt.show()
```

```python
# Buscar el artista mas popular por pais
data["Popularidad"] = pd.to_numeric(data["Popularidad"], errors='coerce')

popularidad_artista_pais = data.groupby(["Pais",'Artista'])["Popularidad"].mean().reset_index()

popularidad_artista_pais
```

```python
artista_mas_popular_pais = popularidad_artista_pais.sort_values(by=["Pais","Popularidad"],ascending=False).drop_duplicates(subset="Pais")
artista_mas_popular_pais
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Pais',y='Popularidad',data=artista_mas_popular_pais,hue="Artista",palette='viridis')
plt.title('Artista mas popular por pais')
plt.xlabel('Pais')
plt.ylabel('Popularidad')
plt.xticks(rotation=45)
plt.show()
```
