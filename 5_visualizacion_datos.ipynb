```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('top_3_latinoamerica_sql.csv')
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Popularidad'],kde=True,bins=30,color="skyblue")
plt.title('Distribucion de popularidad de las canciones')
plt.xlabel('Popularidad')
plt.ylabel('Frecuencia')
plt.show()
```

# Top 3 de canciones con mayour popularidad en latam

```python
top_3 = data.sort_values('Popularidad',ascending=False)
top_3.head(5)
```

```python
top_3 = top_3.drop_duplicates(subset='Nombre de la cancion')
top_3
```

```python
top_3 = top_3.head(3)
top_3
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

bar_plot = sns.barplot(x='Nombre de la cancion',y='Popularidad',data=top_3,palette='viridis')
plt.title('Top 3 canciones mas populares')
plt.xlabel('Nombre de la cancion')
plt.ylabel('Popularidad')
plt.show()
```

# Popularidad promedio en cada pais

```python
poularidad_por_pais = data.groupby('Pais')['Popularidad'].mean().sort_values(ascending=False)
poularidad_por_pais
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
#bar_plot = sns.barplot(x=poularidad_por_pais.index,y=poularidad_por_pais.values,palette='viridis')
bar_plot = sns.barplot(x=poularidad_por_pais.values, y=poularidad_por_pais.index,palette='viridis')
plt.title('Popularidad promedio por pais')
plt.ylabel('Pais')
plt.xlabel('Popularidad promedio')
plt.show()
```

```python
artistas_conteo = data["Artista"].value_counts().reset_index()
artistas_conteo.columns = ["Artista", "Numero de Canciones"]

top_5_artistas = artistas_conteo.head(10)
top_5_artistas
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x="Numero de Canciones",y="Artista",data=top_5_artistas,palette='viridis')
plt.title('Top 5 artistas con mas canciones')
plt.xlabel('Numero de canciones')
plt.ylabel('Artista')
plt.show()
```
